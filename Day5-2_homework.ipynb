```python
#使用 Request 抓取資料
## 假如我們不想把資料載到自己的電腦裡?
# 把連結填入
target_url = ' https://raw.githubusercontent.com/vashineyu/slides_and_others/master/tutorial/examples/imagenet_urls_examples.txt'
```

```python
import requests
response = requests.get(target_url)
txt_data = response.text
txt_data
```

```python
# 用 request 傳送回來的資料不會認得斷行符號
print(len(txt_data))
txt_data[0:100]
```

```python
# 找到換行符號，用該符號做字串分割後，把它拿掉
split_tag = '\t'

data = txt_data.split(split_tag)
print(len(data))
data[0]
```

```python
import pandas as pd

data = []
with open("imagenet_urls_examples.txt", 'r') as f:
    for line in f:
        line = line.replace('\n', '').split('\t') # 將每句最後的 /n 取代成空值後，再以逗號斷句
        data.append(line)
data[0:5]

df = pd.DataFrame(data[0:])
df.columns = ['image_names','image_url']
df[1:]


```
